```python
import pandas as pd
import numpy as np
from nltk.corpus import brown
import pyLDAvis
import pyLDAvis.sklearn
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
```

```python
data = pd.read_csv(r"E:\EEG DATA\train_ordered\02_tcp_le\Text_data\text_meta.csv")
data.head()
```

```python
text = data['report']
text = text.apply(lambda txt: str(txt))
```

```python
seiz = data[data['label'] == 1]
nonseiz = data[data['label'] == 0]
```

```python
vectorizer = CountVectorizer(min_df=3, max_df=0.9, stop_words='english', lowercase=True, token_pattern='[a-zA-Z\-][a-zA-Z\-]{2,}')
lda_model = LatentDirichletAllocation(n_topics=10, max_iter=10, learning_method='online')
```

```python
data_seiz_vectorized = vectorizer.fit_transform(seiz['report'])
lda_seiz = lda_model.fit_transform(data_seiz_vectorized)

pyLDAvis.enable_notebook()
panel = pyLDAvis.sklearn.prepare(lda_model, data_seiz_vectorized, vectorizer, mds='tsne')
panel
```

```python
data_non_vectorized = vectorizer.fit_transform(nonseiz['report'])
lda_non = lda_model.fit_transform(data_non_vectorized)

pyLDAvis.enable_notebook()
panel = pyLDAvis.sklearn.prepare(lda_model, data_non_vectorized, vectorizer, mds='tsne')
panel
```
